```python
import pandas as pd

file = open('cbs_voorlopig.csv')
cbs_df = pd.read_csv(file)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'gem_woz' is numeric
cbs_df['gem_woz'] = pd.to_numeric(cbs_df['gem_woz'], errors='coerce')

# Filter the dataset for "Oud Hoograven-Zuid"
hoograven_Z = cbs_df[cbs_df['regio'] == 'Oud Hoograven-Zuid']

# Plot WOZ values over time
plt.figure(figsize=(10, 6))
plt.plot(hoograven_Z['year'], hoograven_Z['gem_woz'], marker='o', linestyle='-', color='b', label="WOZ Value")

# Add labels and title
plt.title("WOZ Value from Oud Hooggraven-Zuid (2013-2022)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("WOZ Value (in thousands)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
