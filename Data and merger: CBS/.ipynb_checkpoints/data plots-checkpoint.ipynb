```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def dataplot():

    years = []
    n_gemeenten = []
    n_buurten = []
    n_wijken = []


    for year in range(14, 25):
        year = str(year).zfill(2)  # Ensure year is two digits
        year_full = f"20{year}"
        file = f'kwb-{year_full}.xls'

        years.append(year)
        
        try:
            # Load dataset
            df_cbs = pd.read_excel(file, usecols=lambda x: any(sub in x.lower().strip() for sub in ['recs']))

            # Ensure 'recs' column exists
            if 'recs' in df_cbs.columns:
                df_cbs['recs'] = df_cbs['recs'].map(lambda x: x.lower().strip() if isinstance(x, str) else x)

            # Count occurrences of each category
            gemeenten = np.sum(df_cbs['recs'] == 'gemeente')
            buurten = np.sum(df_cbs['recs'] == 'buurt')
            wijken = np.sum(df_cbs['recs'] == 'wijk')

            # Append counts
            n_gemeenten.append(gemeenten)
            n_buurten.append(buurten)
            n_wijken.append(wijken)

    #     except FileNotFoundError:
    #         print(f"File not found: {file}")
    #         n_gemeenten.append(0)
    #         n_buurten.append(0)
    #         n_wijken.append(0)
        except Exception as e:
            print(f"Error processing {file}: {e}")
            n_gemeenten.append(0)
            n_buurten.append(0)
            n_wijken.append(0)

    # Plot results
    plt.figure(figsize=(10, 6))
    plt.plot(years, n_buurten, marker='s', label="Buurten", linestyle='--')
    plt.plot(years, n_wijken, marker='^', label="Wijken", linestyle=':')
    plt.plot(years, n_gemeenten, marker='o', label="Gemeenten", linestyle='-')

    plt.xlabel("Year")
    plt.ylabel("Count (log scale)")
    plt.yscale("log")  # Apply logarithmic scale to y-axis
    plt.title("Count of Gemeenten, Buurten, and Wijken Over the Years (Log Scale)")
    plt.legend()
    plt.grid(True, which="both", linestyle="--")  # Grid for both major and minor ticks
    
    plt.show()

# Run the function
dataplot()

                
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

def dataplot():
    years = []
    n_gemeenten = []
    n_buurten = []
    n_wijken = []

    for year in range(14, 25):
        year = str(year).zfill(2)  # Ensure year is two digits
        year_full = f"20{year}"
        file = f'kwb-{year_full}.xls'

        years.append(int(year_full))  # Convert to full year format (e.g., 2014)

        try:
            # Load dataset
            df_cbs = pd.read_excel(file, usecols=lambda x: 'recs' in x.lower().strip())

            # Ensure 'recs' column exists
            if 'recs' in df_cbs.columns:
                df_cbs['recs'] = df_cbs['recs'].map(lambda x: x.lower().strip() if isinstance(x, str) else x)

            # Count occurrences of each category
            gemeenten = np.sum(df_cbs['recs'] == 'gemeente')
            buurten = np.sum(df_cbs['recs'] == 'buurt')
            wijken = np.sum(df_cbs['recs'] == 'wijk')

            # Append counts
            n_gemeenten.append(gemeenten)
            n_buurten.append(buurten)
            n_wijken.append(wijken)

        except Exception as e:
            print(f"Error processing {file}: {e}")
            n_gemeenten.append(0)
            n_buurten.append(0)
            n_wijken.append(0)

    # Create a figure with 3 side-by-side subplots (no shared y-axis)
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))  

    # Gemeenten Plot
    axes[0].plot(years, n_gemeenten, marker='o', linestyle='-', color='darkslateblue', label="Municipalities")
    axes[0].set_xlabel("Year")
    axes[0].set_ylabel("Count")
    axes[0].set_title("Municipalities Over the Years")
    axes[0].legend()
    axes[0].grid(True)

    # Buurten Plot
    axes[1].plot(years, n_wijken, marker='s', linestyle='--', color='darkslateblue', label="Districts")
    axes[1].set_xlabel("Year")
    axes[1].set_title("Districts Over the Years")
    axes[1].legend()
    axes[1].grid(True)

    # Wijken Plot
    axes[2].plot(years, n_buurten, marker='^', linestyle=':', color='darkslateblue', label="Neighbourhoods")
    axes[2].set_xlabel("Year")
    axes[2].set_title("Neighbourhoods Over the Years")
    axes[2].legend()
    axes[2].grid(True)

    # Adjust layout
    plt.tight_layout()

    # Save plot in the same directory
    save_path = os.path.join(os.getcwd(), "dataplot.png")
    plt.savefig(save_path, dpi=300)
    print(f"Plot saved as {save_path}")

    # Show the figure
    plt.show()

# Run the function
dataplot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

def dataplot():
    years = []
    n_gemeenten = []
    n_buurten = []
    n_wijken = []

    for year in range(14, 16):
        year = str(year).zfill(2)  # Ensure year is two digits
        year_full = f"20{year}"
        file = f'kwb-{year_full}.xls'

        years.append(int(year_full))  # Convert to full year format (e.g., 2014)

        try:
            # Load dataset
            df_cbs = pd.read_excel(file, usecols=lambda x: 'recs' in x.lower().strip())

            # Ensure 'recs' column exists
            if 'recs' in df_cbs.columns:
                df_cbs['recs'] = df_cbs['recs'].map(lambda x: x.lower().strip() if isinstance(x, str) else x)

            # Count occurrences of each category
            gemeenten = np.sum(df_cbs['recs'] == 'gemeente')
            buurten = np.sum(df_cbs['recs'] == 'buurt')
            wijken = np.sum(df_cbs['recs'] == 'wijk')

            # Append counts
            n_gemeenten.append(gemeenten)
            n_buurten.append(buurten)
            n_wijken.append(wijken)

        except Exception as e:
            print(f"Error processing {file}: {e}")
            n_gemeenten.append(0)
            n_buurten.append(0)
            n_wijken.append(0)

    # Create a figure with 3 side-by-side subplots (no shared y-axis)
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))  

    # Gemeenten Plot
    axes[0].plot(years, n_gemeenten, marker='o', linestyle='-', color='darkslateblue', label="Municipalities")
    axes[0].set_xlabel("Year")
    axes[0].set_ylabel("Count")
    axes[0].set_title("Municipalities Over the Years")
    axes[0].legend()
    axes[0].grid(True)

    # Buurten Plot
    axes[1].plot(years, n_wijken, marker='s', linestyle='--', color='deepskyblue', label="Districts")
    axes[1].set_ylabel("Count")  # Each plot has its own y-label now
    axes[1].set_title("Districts Over the Years")
    axes[1].legend()
    axes[1].grid(True)

    # Wijken Plot
    axes[2].plot(years, n_buurten, marker='^', linestyle=':', color='midnightblue', label="Neighbourhoods")
    axes[2].set_ylabel("Count")  # Each plot has its own y-label now
    axes[2].set_title("Neighbourhoods Over the Years")
    axes[2].legend()
    axes[2].grid(True)

    # Adjust layout
    plt.tight_layout()

    # Save plot in the same directory
    save_path = os.path.join(os.getcwd(), "dataplot.png")
    plt.savefig(save_path, dpi=300)
    print(f"Plot saved as {save_path}")

    # Show the figure
    plt.show()

# Run the function
dataplot()
```
