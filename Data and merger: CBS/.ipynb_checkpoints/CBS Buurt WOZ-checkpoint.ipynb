```python
def fill_csv_with_excel(file_path: str, excel_path: str):
    # Load the initialized CSV
    df_csv = pd.read_csv(file_path)
    
    # Load the Excel file
    df_excel = pd.read_excel(excel_path)
    
    # Assuming `df_excel` has columns 'regio', 'gm_naam', 'RECS', 'gwb_code', and other WOZ-related columns
    # Map the columns accordingly (adjust based on your actual column names)
    mapping = {
        "regio": "regio",
        "gm_naam": "gm_naam",
        "recs": "recs",
        "gwb_code": "gwb_code",
    }
    
    # Filter and rename the columns from Excel data
    df_to_append = df_excel.rename(columns=mapping)[mapping.values()]
    
    # Append new data to the CSV DataFrame
    df_csv = pd.concat([df_csv, df_to_append], ignore_index=True)
    
    # Save back to CSV
    df_csv.to_csv(file_path, index=False)
    print(f"CSV updated at: {file_path}")

# Paths
file_path = '/home/wouter/Documents/Scriptie/datacbs/cbs_woz.csv'
excel_path = 'kwb-2024.xls'

# Initialize the CSV
initialize_csv_cbs(file_path)

# Fill the CSV with Excel data
fill_csv_with_excel(file_path, excel_path)
```

```python
import pandas as pd

def cbs(path_file_to):
    cbs_woz_path = path_file_to
    df_woz = pd.read_csv(cbs_woz_path, low_memory=False)

    mismatches = {}  # Dictionary to track mismatches

    for year in range(12, 25):
        year = str(year).zfill(2)  # Ensure year is two digits
        year_full = f"20{year}"
        file = f'kwb-{year_full}.xls'

        try:
            df_cbs = pd.read_excel(file)
        except Exception as e:
            print(f"Could not open {file}: {e}")
            continue

        df_cbs.columns = df_cbs.columns.str.strip().str.lower()
        df_woz.columns = df_woz.columns.str.strip().str.lower()

        woz_col = [col for col in df_cbs.columns if 'woz' in col or col == 'woningwaarde']
        woz_col = woz_col[0]
        #print(woz_col)

        
        if 'regio' in df_cbs.columns and 'gwb_code' in df_cbs.columns: 
            print(f"Processing WOZ columns for year 20{year}")

            df_woz = df_woz.merge(
                df_cbs[['regio', 'recs', 'gm_naam', woz_col]],  # Select the relevant columns from df_cbs
                left_on=['regio', 'recs', 'gm_naam'],  # Columns from df_woz
                right_on=['regio', 'recs', 'gm_naam'],  # Columns from df_cbs
                how='left'  # Perform a left join
            )
            
            df_woz[f'gem_woz_{year_full}'] = df_cbs[woz_col]
            df_woz.drop(columns=[woz_col], inplace=True)
            

        #elif f'gwb_naam{year}_60pos' in df_cbs.columns and f'gwb_code{year}' in df_cbs.columns:
        elif f'gwb_naam{year}_60pos' in df_cbs.columns and 'gm_naam' in df_cbs.columns:    
            print(f"Processing WOZ columns for year 20{year}")

            #print(df_cbs.columns[0:6])

            df_woz = df_woz.merge(
                df_cbs[[f'gwb_naam{year}_60pos', 'gm_naam', woz_col]],  # Select the relevant columns from df_cbs
                left_on=['regio', 'gm_naam'],  # Columns from df_woz
                right_on=[f'gwb_naam{year}_60pos', 'gm_naam'],  # Columns from df_cbs
                how='left'  # Perform a left join
            )

            df_woz[f'gem_woz_{year_full}'] = df_cbs[woz_col]
            df_woz.drop(columns=[woz_col, f'gwb_naam{year}_60pos'], inplace=True)
                
            
        df_woz.to_csv(cbs_woz_path, index=False)
            
        

# Call the function and display the DataFrame
df = cbs('/home/wouter/Documents/Scriptie/datacbs/cbs_woz.csv')
#print(df.head())
```
