```python
import pandas as pd
from rapidfuzz import fuzz  # For fuzzy matching

def compare_differences(df_old, df_new, name_col, code_col, gm_col, typ_R):
    matches = []
    missing = []
    second_time_match = []

    # Convert df_old names to lowercase for case-insensitive comparison
    df_old[name_col] = df_old[name_col].fillna("").str.lower()
    df_old[gm_col] = df_old[gm_col].fillna("").str.lower()
    df_old[typ_R] = df_old[typ_R].fillna("").str.lower()


    # Create a set of tuples from old data for fast lookup
    df_old_set_name = set(zip(df_old[name_col], df_old[gm_col], df_old[typ_R]))

    # Iterate through df_new rows
    for index, row in df_new.iterrows():
        key_name = (row['regio'].lower(), row['gm_naam'].lower(), row['recs'].lower())
        
        if key_name in df_old_set_name:
            matches.append({
                "index": index,
                "regio": row['regio'],
                "gm_naam": row['gm_naam'],
                "note": "Exact match found"
            })
        else:
            # Fuzzy matching
            best_score = 0
            best_match = None

            for _, row_old in df_old.iterrows():
                if row_old[gm_col] == row['gm_naam'].lower() and row_old[typ_R] == row['recs'].lower():
                    similarity = fuzz.ratio(row['regio'].lower(), row_old[name_col].lower())

                    if similarity > best_score and similarity < 100:
                        best_score = similarity
                        best_match = row_old[name_col]

            
            # Log mismatches based on a threshold
            if best_score > 70:
                print(f"GM {row['gm_naam']} - Old name: {best_match}, Resembles New one: {row['regio'].lower()}, with score: {best_score}")
                second_time_match.append({
                    "index": index,
                    "gm_naam": row['gm_naam'],
                    "NEW regio": row['regio'],
                    "OLD regio": best_match,
                    "Simalarity": best_score

                })

            
            else:
                missing.append({
                    "index": index,
                    "regio": row['regio'],
                    "gm_naam": row['gm_naam'],
                    "note": "No exact match found"
                })

    return matches, missing, second_time_match
```

```python
def initialize_csv_cbs(file_path: str):
    # Define the columns as per the specified labels
    columns = [
        "regio", "gm_naam", "recs", "gwb_code", "year", "gem_woz"
    ]
    # Create an empty DataFrame with these columns
    df = pd.DataFrame(columns=columns)
    # Save to CSV
    df.to_csv(file_path, index=False)
    print(f"Initialized CSV at: {file_path}")
```

```python
def finalize_csv_format(file_path: str):
    """
    Finalize the CSV file by grouping rows and sorting them by gm_naam, gwb_code, and year.
    """
    try:
        # Load the CSV
        df = pd.read_csv(file_path)
        
        # Sort the DataFrame by gm_naam, gwb_code, and year
        df = df.sort_values(by=["gm_naam", "gwb_code", "year"], ascending=[True, True, True])
        
        # Save back to the same CSV file
        df.to_csv(file_path, index=False)
        print(f"CSV file '{file_path}' has been formatted and saved.")
    except Exception as e:
        print(f"Error during finalizing CSV: {e}")
```

```python
def save_matches_to_csv(matches, file_path):
    # Create or append to a CSV file with matches
    df_matches = pd.DataFrame(matches)
    
    if not df_matches.empty:
        df_matches.to_csv(file_path, mode='a', index=False, header=False)
        print(f"Appended {len(matches)} matches to {file_path}")
```
