```python
def create_time_woz_features(df):

    # Create lag features for WOZ
    df["woz_lag_1"] = df.groupby("gwb_code")["woz"].shift(1)
    df["woz_lag_2"] = df.groupby("gwb_code")["woz"].shift(2)
    df["woz_lag_3"] = df.groupby("gwb_code")["woz"].shift(3)
    df["woz_lag_4"] = df.groupby("gwb_code")["woz"].shift(3)
    df["woz_lag_5"] = df.groupby("gwb_code")["woz"].shift(3)

    # Create rolling average features
    df["woz_rolling_avg_3"] = df.groupby("gwb_code")["woz"].transform(lambda x: x.shift(1).rolling(3).mean())

    return df
    
train_df_mv = create_time_woz_features(df_selected)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

def MICE_with_lag(df):    
    # Define MICE imputer (uses regression models to fill missing values)
    mice_imputer = IterativeImputer(max_iter=10, random_state=42)

    # Step 1: Scale before imputation
    scaler = StandardScaler()
    scaled_features = ["woz_rolling_avg_3", "woz_lag_1", "woz_lag_2", "woz_lag_3", "woz_lag_4", "woz_lag_5"]
    df[scaled_features] = scaler.fit_transform(df[scaled_features])
    
    # Step 2: Apply MICE imputation
    lag_features = scaled_features + ["year"]
    df[lag_features] = mice_imputer.fit_transform(df[lag_features])
    
    # Step 3: Reverse the scaling (convert back to original values)
    df[scaled_features] = scaler.inverse_transform(df[scaled_features])

    return df

# def MICE_no_lag(df):    
#     # Define MICE imputer (uses regression models to fill missing values)
#     mice_imputer = IterativeImputer(max_iter=10, random_state=42)

#     feature_cols = list(df.columns)  # Convert Index to a list of column names
    
#     scaler = StandardScaler()
#     scaled_features = [col for col in feature_cols if col not in ["woz", "year", "gwb_code", "gm_naam"]]
#     df[scaled_features] = scaler.fit_transform(df[scaled_features])

#     imput_features = scaled_features + ["year"] 
#     df[imput_features] = mice_imputer.fit_transform(df[imput_features])

#     df[scaled_features] = scaler.inverse_transform(df[scaled_features])


#     return df

def ffil(df):
    """
    Forward-fills missing values column-wise.
    
    If a column has missing values, it fills them using forward fill (ffill).
    If ffill doesn't work for the first row(s), it backfills (bfill) as a fallback.
    """
    df = df.copy()  # Ensure we don’t modify the original DataFrame
    for col in df.columns:
        if df[col].isnull().any():  # Check if the column has missing values
            df[col] = df[col].ffill().bfill()  # Apply forward fill, then backfill as fallback

    return df

```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from xgboost import plot_importance
import matplotlib.pyplot as plt

def split_data(df, test_year):
    train_df = df[df['year'] < test_year]
    test_df = df[df['year'] == test_year]  # Use 2024 as test year
    
    # Define features and target
    X_train = train_df.drop(columns=['woz'])
    y_train = train_df['woz']
    X_test = test_df.drop(columns=['woz'])
    y_test = test_df['woz']

    return X_train, X_test, y_train, y_test

def evaluate_model(y_test, y_pred):
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Print results
    print("Mean Squared Error (MSE):", mse)
    print("R-squared (R2):", r2)
    
    # Display predictions alongside actual values
    results = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
    print("\nPredictions vs Actual:")
    print(results)

    return results, mse, r2

def train_model(df):

    model = XGBRegressor(n_estimators=500, learning_rate=0.1, max_depth=5, random_state=42)

    # Train/Test Split
    X_train, X_test, y_train, y_test = split_data(df, 2024)

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    result_XGB, mse_XGB, r2_XGB = evaluate_model(y_test, y_pred)

    
    feature_importance = model.feature_importances_

    importance_df = pd.DataFrame({
        "Feature": X_train.columns,
        "Importance": feature_importance
    })
    
    importance_df = importance_df.sort_values(by="Importance", ascending=False)
    
    # Select the top 10 most important features
    top_10_features = importance_df.head(10)
    
    # Display the top 10 features
    print("\nTop 10 Most Important Features:")
    print(top_10_features)
    
    # Plot the top 10 most important features
    plt.figure(figsize=(10, 6))
    plt.barh(top_10_features["Feature"], top_10_features["Importance"])
    plt.xlabel("Feature Importance")
    plt.ylabel("Features")
    plt.title("Top 10 Most Important Features - XGBoost")
    plt.gca().invert_yaxis()  # Show most important feature at the top
    plt.show()

    return model
model = train_model(df_train)
#model = train_model(train_df_no_mv)
```

```python
def train_2025(df_train, df_for_pred, year):
    pred_df = df_for_pred.copy()

    # Train het model op de volledige dataset (inclusief 2024)
    X_full = df_train.drop(columns=["woz"])
    y_full = df_train["woz"]
    
    model = XGBRegressor(n_estimators=500, learning_rate=0.1, max_depth=5, random_state=42)
    model.fit(X_full, y_full)

    #drop the lags and rolling mean in df_for_pred
    pred_df_no_lag = pred_df.drop(columns=["woz_rolling_avg_3", "woz_lag_1", "woz_lag_2", "woz_lag_3", "woz_lag_4", "woz_lag_5"])

    # impute the 2025 missing values
    pred_df_no_lag = ffil(pred_df_no_lag)
    
    # Give them lags, time features and handle the MV's
    pred_df_lag = create_time_woz_features(pred_df_no_lag)
    pred_df_lag_no_mv = MICE_with_lag(pred_df_lag)
    
    #print(pred_df_lag_no_mv)
    # Gebruik de 2024-features om 2025 te voorspellen
    X_25 = pred_df_lag_no_mv[pred_df_lag_no_mv["year"] == 2025][X_full.columns]  # Neem 2024 als input voor 2025
    
    #print(X_25)

    # Predict 2025 WOZ values
    y_25 = model.predict(X_25)

    #put the predicted values back
    pred_df_lag_no_mv.loc[pred_df_lag_no_mv["year"] == 2025, "woz"] = y_25

    # Return the updated DataFrame
    return pred_df_lag_no_mv, model, y_25, X_full

# train_df_no_mv, is complete with 24 data, including lags
# pred_df is empty for rows 2025 data
df_25_with_lag, model, y_25, X_full = train_2025(train_df_no_mv, pred_df, 2025)
```
