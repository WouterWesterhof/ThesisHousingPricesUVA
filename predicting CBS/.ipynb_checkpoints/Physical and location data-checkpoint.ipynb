## OpenStreetMap

```python
import osmnx as ox
from IPython.display import Image
import networkx as nx
import folium
```

```python
# configure the inline image display
img_folder = "images"
extension = "png"
size = 240
```

```python

# place = {
#         "city": "Utrecht",
#         "province": "Utrecht",
#         "country": "Netherlands",
#     }
ox.settings.overpass_endpoint = "https://overpass.kumi.systems/api/interpreter"  # Alternative API

place = {
        "city": "Utrecht",
        "province": "Utrecht",
        "country": "Netherlands",
    }

tags = {'building': True}

# Fetch building footprints
buildings = ox.features_from_place(place, tags=tags)
buildings = buildings[buildings.geometry.geom_type == 'Polygon']

# Plot the building footprints
ax = buildings.plot(figsize=(6, 6), color = 'orange')
ax.set_title('Building Footprints in Maastricht, Netherlands')
```

```python
gdf = ox.features.features_from_place("Piedmont, California, USA", tags)
gdf_proj = ox.projection.project_gdf(gdf)
fp = f"./{img_folder}/piedmont_bldgs.{extension}"
fig, ax = ox.plot.plot_footprints(gdf_proj, filepath=fp, dpi=400, save=True, show=False, close=True)
Image(fp, height=size, width=size)
```
