```python
from __future__ import absolute_import, division, print_function, \
    with_statement
import requests
from peewee import * 
import json
import numpy
import logging
from multiprocessing.dummy import Pool as ThreadPool
import threading
```

```python
def save_to_list(json_string):
    obj_list = []
    try:
        # Ensure the response is not empty before parsing
        if not json_string.strip():
            raise ValueError("Empty JSON string received.")
        
        json_obj = json.loads(json_string)
        obj_list.append(json_obj)
        print(obj_list)
        return obj_list
    
    except json.JSONDecodeError as e:
        logging.error("JSON parsing failed: %s", e)
        logging.error("Response content: %s", json_string)
        return None
```

```python
def scrape_range_and_save(args):
    global objects_total, objects_done
    json_string = scrape_obj_from_id_to_id(args[0], args[1])
    print("----------------- Databased dumped, start parsing :) -------------------", end="\n")
    save_to_list(json_string=json_string)
    objects_done += 1
    print("----------------- Process: {:2.2f}%------------------".format((objects_done)*100/(objects_total)))
```
