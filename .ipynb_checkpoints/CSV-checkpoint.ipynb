## Extend the df with WOZ-values

To predict the increase of the house prices in the Netherlands, is it necessary to get an insight of the houseprices in the past. The value (expressed as the WOZ) of all the houses can be found on `https://www.wozwaardeloket.nl/`.

So what is this data all about, lets check using `IJzerdraadpad 1` as adress.

```python
from wozpy.wozpy import Wozpy

woz = Wozpy()


woz_value = woz.get_woz_value("IJzerdraadpad 33 3525BP")

for line in woz_value:
    for bom in line:
        print(bom)
```

We see some interesting features we want to include in our dataset, those are the **WOZ-values** from 2014 until 2023. But also features as 'Bouwjaar', 'Oppervlakte' en 'Grondoppervlakte'. So the function `initialize_csv` creates .csv file that includes those features, where the data will be stored.

```python
def initialize_csv(file_path: str):
    # Define the columns as per the specified labels
    columns = [
        "address", "bouwjaar", "grondoppervlakte", "oppervlakte",
        "woz_2014", "woz_2015", "woz_2016", "woz_2017", "woz_2018",
        "woz_2019", "woz_2020", "woz_2021", "woz_2022", "woz_2023"
    ]

    # Create an empty DataFrame with these columns
    df = pd.DataFrame(columns=columns)

    # Save to CSV
    df.to_csv(file_path, index=False)
    print(f"Initialized CSV at: {file_path}")

initialize_csv('/home/wouter/Documents/Scriptie/csv/property_data.csv')
```

The data that will be stored using the class `Property`, each address and their relevant features will be stored in 'property_data.csv'.

```python
import pandas as pd
import requests
from typing import Optional, Dict, List
import os
import random
import time

class Property:
    def __init__(self, address: str, postcode: str, woonplaats: str):
        self.address = address  
        self.postcode = postcode
        self.woonplaats = woonplaats

    def get_address_ids(self, complete_address):
        url = "https://api.pdok.nl/bzk/locatieserver/search/v3_1/suggest"
        params = {"q": complete_address}

        try:
            response = requests.get(url, params=params, timeout=5)
            response.raise_for_status()
            data = response.json()
            return [doc.get("id", "") for doc in data.get("response", {}).get("docs", [])]
        except (requests.RequestException, ValueError, KeyError, IndexError) as ex:
            print(f"Error while retrieving or parsing address ID: {ex}")
            return None

    def get_correct_addressID(self, complete_address):
        _ids = self.get_address_ids(complete_address)
        url = "https://api.pdok.nl/bzk/locatieserver/search/v3_1/lookup"

        nummer_and_adresseer_ids = {'nummeraanduiding_ids': [], 'adresseerbaarobject_ids': []}

        for _id in _ids:
            data = self.data_for_adressID(url, _id)
            nummeraanduiding_id = (
                data.get("response", {}).get("docs", [{}])[0].get("nummeraanduiding_id")
            )
            adresseerbaarobject_id = (
                data.get("response", {}).get("docs", [{}])[0].get("adresseerbaarobject_id")
            )

            if nummeraanduiding_id and adresseerbaarobject_id:
                nummer_and_adresseer_ids["nummeraanduiding_ids"].append(nummeraanduiding_id)  
                nummer_and_adresseer_ids["adresseerbaarobject_ids"].append(adresseerbaarobject_id)
        return nummer_and_adresseer_ids

    def data_for_adressID(self, url: str, _id: str) -> dict:
        params = {"id": _id}
        try:
            response = requests.get(url=url, params=params, timeout=5)
            response.raise_for_status()
            return response.json()
        except requests.RequestException as e:
            print(f"Request failed for {self.address} with error: {e}")
            return {}

    def get_woz_values(self, _ids: List[str]) -> Dict[int, Optional[int]]:
        headers = {
            'sec-ch-ua-platform': '"Linux"',
            'Referer': '',
            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',
            'Accept': 'application/json, text/plain, */*',
            'sec-ch-ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
            'sec-ch-ua-mobile': '?0',
        }
        
        base_url = "https://api.kadaster.nl/lvwoz/wozwaardeloket-api/v1/wozwaarde/nummeraanduiding/"
        woz_values = {year: None for year in range(2014, 2024)}  # 2014 to 2023
        grondoppervlakte = None
        max_retries = 4
        
        for _id in _ids:
            attempt = 0
            while attempt < max_retries:
                try:
                    response = requests.get(f"{base_url}{_id}", timeout=5)
                    if response.status_code == 429:
                        wait_time = 2 ** attempt + random.uniform(0, 1)  # Adds jitter
                        print(f"Rate limited. Waiting for {wait_time:.2f} seconds...")
                        time.sleep(wait_time)
                        attempt += 1
                        continue
                    
                        

                    response.raise_for_status()  # Raises HTTPError if status is 4xx or 5xx
                    data = response.json()

                    
                    for item in data['wozWaarden']:
                        year = int(item['peildatum'][:4])
                        if 2014 <= year <= 2023:
                            woz_values[year] = item['vastgesteldeWaarde']
                
                    grondoppervlakte = data['wozObject'].get('grondoppervlakte', grondoppervlakte)
                    break  # Exit retry loop if successful
                    
                except requests.exceptions.HTTPError as e:
                    print(f"Error retrieving WOZ data for adresss: {self.address} with {_id}: {e}")
                    break  # Exit on HTTP errors other than 429
                
                # **Added code to handle network-related retries**
                except (requests.exceptions.ConnectionError, requests.exceptions.Timeout) as e:
                    print(f"Network issue for {_id}. Attempt {attempt + 1}/{max_retries}. Retrying in 2 seconds...")
                    time.sleep(2)  # Wait before retrying
                    attempt += 1
                    continue  # Retry for network errors
    
                # **Exit if it's a non-recoverable error, like a general request exception**
                except requests.exceptions.RequestException as e:
                    print(f"Request failed for {self.address} with error: {e}")
                    break 
                    
            else:
                print(f"Failed to retrieve WOZ data for {_id} after {max_retries} retries.")

        return woz_values, grondoppervlakte

    def get_other_features(self, _ids: List[str]):
        headers = {
            'Accept': 'application/json, text/plain, */*',
            'User-Agent': 'Mozilla/5.0',
        }
        params = {'service': 'WFS'}
        features = {}
    
        for _id in _ids:
            data = (
                f'<wfs:GetFeature xmlns:wfs="http://www.opengis.net/wfs" '
                f'service="WFS" version="1.1.0" xsi:schemaLocation="http://www.opengis.net/wfs '
                f'http://schemas.opengis.net/wfs/1.0.0/WFS-transaction.xsd" '
                f'xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" outputFormat="application/json">'
                f'<wfs:Query typeName="verblijfsobject" xmlns:null="http://bag.geonovum.nl">'
                f'<ogc:Filter xmlns:ogc="http://www.opengis.net/ogc">'
                f'<ogc:PropertyIsEqualTo>'
                f'<ogc:PropertyName>identificatie</ogc:PropertyName>'
                f'<ogc:Literal>0{int(_id)}</ogc:Literal>'
                f'</ogc:PropertyIsEqualTo>'
                f'</ogc:Filter>'
                f'</wfs:Query>'
                f'</wfs:GetFeature>'
            )
            try:
                response = requests.post(
                    'https://service.pdok.nl/lv/bag/wfs/v2_0',
                    params=params,
                    headers=headers,
                    data=data,
                    timeout=5
                )
                response.raise_for_status()
    
                # Check if response content type is JSON
                if response.headers.get("Content-Type") == "application/json":
                    data = response.json()
                else:
                    print(f"Unexpected content type: {response.headers.get('Content-Type')}")
                    print("Response text:", response.text)
                    continue  # Skip this ID if the content is not JSON
    
                # Ensure data is in expected format
                if 'features' in data and data['features']:
                    feature_data = data['features'][0]['properties']
                    features['gebruiksdoel'] = feature_data.get('gebruiksdoel')
                    features['bouwjaar'] = feature_data.get('bouwjaar')
                    features['oppervlakte'] = feature_data.get('oppervlakte')
                else:
                    print(f"No features found for ID {_id}")
            except requests.exceptions.RequestException as ex:
                print(f"Request failed for ID {self.address} with error: {ex}")
            except ValueError as json_ex:
                print(f"Failed to parse JSON for ID {_id}: {json_ex}")
    
        return features


    def to_dataframe(self) -> pd.DataFrame:
        complete_address = f'{self.postcode} {self.address} {self.woonplaats}'
        ids = self.get_correct_addressID(complete_address)
        
        if not ids:
            return pd.DataFrame()  # Empty DataFrame if no IDs

        # Other features
        other_features = self.get_other_features(ids["adresseerbaarobject_ids"])

        # Check if "gebruiksdoel" is "woonfunctie"; if not, skip this property
        if other_features.get("gebruiksdoel") != "woonfunctie":
            print(f"Skipping property at {self.address} as 'gebruiksdoel' is not 'woonfunctie'")
            return pd.DataFrame()  # Return an empty DataFrame to indicate skipping

        # WOZ values from 2014 to 2023
        woz_values, grondoppervlakte = self.get_woz_values(ids["nummeraanduiding_ids"])
        
        # Prepare data for DataFrame
        data = {
            "address": self.address,
            "bouwjaar": other_features.get("bouwjaar"),
            "grondoppervlakte": grondoppervlakte,
            "oppervlakte": other_features.get("oppervlakte"),
            **{f"woz_{year}": woz_values.get(year) for year in range(2014, 2024)},
        }

        self.save_to_csv(pd.DataFrame([data]))

    def save_to_csv(self, data: pd.DataFrame):
        # Specify the file path
        file_path = '/home/wouter/Documents/Scriptie/csv/property_data.csv'
        
        # Append data to CSV if not empty
        if not data.empty:
            data.to_csv(file_path, mode='a', header=not os.path.exists(file_path), index=False)
            print(f"Data for property at {self.address} saved to {file_path}")
        else:
            print(f"No data to save for property at {self.address}")
```

```python
    import pandas as pd
    import requests
    from typing import Optional, Dict, List
    import os
    import random
    import time
    from concurrent.futures import ThreadPoolExecutor, as_completed
    
    
    class Property2:
        def __init__(self, address: str, postcode: str, woonplaats: str):
            self.address = address  
            self.postcode = postcode
            self.woonplaats = woonplaats
    
        def get_address_ids(self, complete_address):
            """
            Retrieve the address ID(s) from the WOZ web service, in this form: ['adr-1b87ea02fd9af9628c06048adaab0eb9'].
    
            Args:
                complete_address (str): The address for which to retrieve WOZ address ID(s).
    
            Returns:
                List[str]]: A list of WOZ address ID(s) if found, otherwise `None`.
            """
            
            url = "https://api.pdok.nl/bzk/locatieserver/search/v3_1/suggest"
            params = {"q": complete_address}
    
            try:
                response = requests.get(url, params=params, timeout=5)
                response.raise_for_status()
                data = response.json()
                return [doc.get("id", "") for doc in data.get("response", {}).get("docs", [])]

            except (requests.RequestException, ValueError, KeyError, IndexError) as ex:
                print(f"Error while retrieving or parsing address ID: {ex}")
                return None
    
        def get_correct_addressID(self, complete_address):
            """
            Retrieve the nummeraanduiding ID(s) as the adresseerbaarobject ID(s) from the WOZ web service.
            To retrieve these ID's the get_address_ids function is needed to obtain the address ID.
            The nummeraanduiding ID is needed to obtain WOZ values.
            The adresseerbaarobject ID is needed to obtain some other features, as bouwjaar en oppervlakte.
            
            Args:
                complete_address (str): The address for which to retrieve WOZ address ID(s).
    
            Returns:
                Dict[str]: A dict containing both "nummeraanduiding_ids" as "adresseerbaarobject_ids" of a given adress.
                Example: {'nummeraanduiding_ids': ['0344200000167932'], 'adresseerbaarobject_ids': ['0344010000165059']}
            """
            _ids = self.get_address_ids(complete_address)
            url = "https://api.pdok.nl/bzk/locatieserver/search/v3_1/lookup"
    
            nummer_and_adresseer_ids = {'nummeraanduiding_ids': [], 'adresseerbaarobject_ids': []}
    
            # Use ThreadPoolExecutor to fetch data concurrently
            with ThreadPoolExecutor() as executor:
                futures = [executor.submit(self.data_for_adressID, url, _id) for _id in _ids]
                for future in as_completed(futures):
                    data = future.result()
                    nummeraanduiding_id = (
                        data.get("response", {}).get("docs", [{}])[0].get("nummeraanduiding_id")
                    )
                    adresseerbaarobject_id = (
                        data.get("response", {}).get("docs", [{}])[0].get("adresseerbaarobject_id")
                    )
    
                    if nummeraanduiding_id and adresseerbaarobject_id:
                        nummer_and_adresseer_ids["nummeraanduiding_ids"].append(nummeraanduiding_id)
                        nummer_and_adresseer_ids["adresseerbaarobject_ids"].append(adresseerbaarobject_id)
            return nummer_and_adresseer_ids
    
    
        def data_for_adressID(self, url: str, _id: str) -> dict:
            params = {"id": _id}
            try:
                response = requests.get(url=url, params=params, timeout=5)
                response.raise_for_status()
                return response.json()
            except requests.RequestException as e:
                print(f"Request failed for {self.address} with error: {e}")
                return {}
    
    
        def get_woz_values(self, _ids: List[str]):
            headers = {
                'sec-ch-ua-platform': '"Linux"',
                'Referer': '',
                'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',
                'Accept': 'application/json, text/plain, */*',
                'sec-ch-ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
                'sec-ch-ua-mobile': '?0',
            }
            
            base_url = "https://api.kadaster.nl/lvwoz/wozwaardeloket-api/v1/wozwaarde/nummeraanduiding/"
            woz_values = {year: None for year in range(2014, 2024)}  # 2014 to 2023
            grondoppervlakte = None
            max_retries = 4
            
            with ThreadPoolExecutor() as executor:
                futures = [executor.submit(self.fetch_woz_value, base_url, _id, woz_values) for _id in _ids]
                for future in as_completed(futures):
                    future.result()  # Ensures completion of the request
            
            return woz_values, grondoppervlakte
    
        def fetch_woz_value(self, base_url, _id, woz_values):
            attempt = 0
            while attempt < 4:
                try:
                    response = requests.get(f"{base_url}{_id}", timeout=5)
                    if response.status_code == 429:
                        wait_time = 2 ** attempt + random.uniform(0, 1)  # Adds jitter
                        print(f"Rate limited. Waiting for {wait_time:.2f} seconds...")
                        time.sleep(wait_time)
                        attempt += 1
                        continue
                    
                    response.raise_for_status()  # Raises HTTPError if status is 4xx or 5xx
                    data = response.json()
                    
                    if 'wozWaarden' not in data or not data['wozWaarden']:
                        print(f"No WOZ values found for ID {_id}, skipping.")
                        return False  # Indicate failure to retrieve data
                
                    for item in data['wozWaarden']:
                        year = int(item['peildatum'][:4])
                        if 2014 <= year <= 2023:
                            woz_values[year] = item['vastgesteldeWaarde']
    
                    grondoppervlakte = data['wozObject'].get('grondoppervlakte', None)
                    return True
                    
                except requests.exceptions.RequestException as e:
                    print(f"Request failed for {self.address} with error: {e} bomadabsdao")
                    return False
    
        def get_other_features(self, _ids: List[str]):
            headers = {
                'Accept': 'application/json, text/plain, */*',
                'User-Agent': 'Mozilla/5.0',
            }
            params = {'service': 'WFS'}
            features = {}
        
            with ThreadPoolExecutor() as executor:
                futures = [executor.submit(self.fetch_other_feature, _id) for _id in _ids]
                for future in as_completed(futures):
                    data = future.result()
                    if data:
                        features.update(data)
            return features
    
        def fetch_other_feature(self, _id: str):
            headers = {
                'Accept': 'application/json, text/plain, */*',
                'User-Agent': 'Mozilla/5.0',
            }
            params = {'service': 'WFS'}
            features = {}
    
            data = (
                f'<wfs:GetFeature xmlns:wfs="http://www.opengis.net/wfs" '
                f'service="WFS" version="1.1.0" xsi:schemaLocation="http://www.opengis.net/wfs '
                f'http://schemas.opengis.net/wfs/1.0.0/WFS-transaction.xsd" '
                f'xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" outputFormat="application/json">'
                f'<wfs:Query typeName="verblijfsobject" xmlns:null="http://bag.geonovum.nl">'
                f'<ogc:Filter xmlns:ogc="http://www.opengis.net/ogc">'
                f'<ogc:PropertyIsEqualTo>'
                f'<ogc:PropertyName>identificatie</ogc:PropertyName>'
                f'<ogc:Literal>0{int(_id)}</ogc:Literal>'
                f'</ogc:PropertyIsEqualTo>'
                f'</ogc:Filter>'
                f'</wfs:Query>'
                f'</wfs:GetFeature>'
            )
    
            try:
                response = requests.post(
                    'https://service.pdok.nl/lv/bag/wfs/v2_0',
                    params=params,
                    headers=headers,
                    data=data,
                    timeout=5
                )
                response.raise_for_status()
    
                if response.headers.get("Content-Type") == "application/json":
                    data = response.json()
                    if 'features' in data and data['features']:
                        feature_data = data['features'][0]['properties']
                        return {
                            'gebruiksdoel': feature_data.get('gebruiksdoel'),
                            'bouwjaar': feature_data.get('bouwjaar'),
                            'oppervlakte': feature_data.get('oppervlakte')
                        }
                    
            except requests.exceptions.RequestException as ex:
                print(f"Request failed for ID {self.address} with error: {ex}, peoepepdepde")
            return {}
    
        def to_dataframe(self):
            complete_address = f'{self.postcode} {self.address} {self.woonplaats}'
            ids = self.get_correct_addressID(complete_address)
            
            if not ids:
                return pd.DataFrame()  # Empty DataFrame if no IDs
    
            # Other features
            other_features = self.get_other_features(ids["adresseerbaarobject_ids"])
    
            # Check if "gebruiksdoel" is "woonfunctie"; if not, skip this property
            if other_features.get("gebruiksdoel") != "woonfunctie":
                print(f"Skipping property at {self.address} as 'gebruiksdoel' is not 'woonfunctie'")
                return pd.DataFrame()  # Return an empty DataFrame to indicate skipping
    
            # WOZ values from 2014 to 2023
            woz_values, grondoppervlakte = self.get_woz_values(ids["nummeraanduiding_ids"])
            
            if all(value is None for value in woz_values.values()) or not self.fetch_woz_value('https://api.kadaster.nl/lvwoz/wozwaardeloket-api/v1/wozwaarde/nummeraanduiding/', ids["nummeraanduiding_ids"][0], woz_values):
                print(f"No WOZ values available for {self.address}, skipping this property.")
                return pd.DataFrame()
            
            # Prepare data for DataFrame
            data = {
                "address": self.address,
                "bouwjaar": other_features.get("bouwjaar"),
                "grondoppervlakte": grondoppervlakte,
                "oppervlakte": other_features.get("oppervlakte"),
                **{f"woz_{year}": woz_values.get(year) for year in range(2014, 2024)},
            }
    
            self.save_to_csv(pd.DataFrame([data]))
    
        def save_to_csv(self, data: pd.DataFrame):
            file_path = '/home/wouter/Documents/Scriptie/csv/property_data.csv'
            
            if not data.empty:
                data.to_csv(file_path, mode='a', header=not os.path.exists(file_path), index=False)
                print(f"Data for property at {self.address} saved to {file_path}")
            else:
                print(f"No data to save for property at {self.address}")
```

```python
class Property3:
    def __init__(self, address: str, postcode: str, woonplaats: str):
        self.address = address  
        self.postcode = postcode
        self.woonplaats = woonplaats

    def get_address_ids(self, complete_address):
        """
        Retrieve the address ID(s) from the WOZ web service, in this form: ['adr-1b87ea02fd9af9628c06048adaab0eb9'].

        Args:
            complete_address (str): The address for which to retrieve WOZ address ID(s).

        Returns:
            List[str]]: A list of WOZ address ID(s) if found, otherwise `None`.
        """
        
        url = "https://api.pdok.nl/bzk/locatieserver/search/v3_1/suggest"
        params = {"q": complete_address}

        try:
            response = requests.get(url, params=params, timeout=5)
            response.raise_for_status()
            data = response.json()
            return [doc.get("id", "") for doc in data.get("response", {}).get("docs", [])]

        except (requests.RequestException, ValueError, KeyError, IndexError) as ex:
            print(f"Error while retrieving or parsing address ID: {ex}")
            return None
    
    def get_correct_addressID(self, complete_address):
        """
        Retrieve the nummeraanduiding ID(s) as the adresseerbaarobject ID(s) from the WOZ web service.
        To retrieve these ID's the get_address_ids function is needed to obtain the address ID.
        The nummeraanduiding ID is needed to obtain WOZ values.
        The adresseerbaarobject ID is needed to obtain some other features, as bouwjaar en oppervlakte.
        
        Args:
            complete_address (str): The address for which to retrieve WOZ address ID(s).

        Returns:
            Dict[str]: A dict containing both "nummeraanduiding_ids" as "adresseerbaarobject_ids" of a given adress.
            Example: {'nummeraanduiding_ids': ['0344200000167932'], 'adresseerbaarobject_ids': ['0344010000165059']}
        """
        _ids = self.get_address_ids(complete_address)
        url = "https://api.pdok.nl/bzk/locatieserver/search/v3_1/lookup"

        nummer_and_adresseer_ids = {'nummeraanduiding_ids': [], 'adresseerbaarobject_ids': []}

        # Use ThreadPoolExecutor to fetch data concurrently
        with ThreadPoolExecutor() as executor:
            futures = [executor.submit(self.data_for_adressID, url, _id) for _id in _ids]
            for future in as_completed(futures):
                data = future.result()
                nummeraanduiding_id = (
                    data.get("response", {}).get("docs", [{}])[0].get("nummeraanduiding_id")
                )
                adresseerbaarobject_id = (
                    data.get("response", {}).get("docs", [{}])[0].get("adresseerbaarobject_id")
                )

                if nummeraanduiding_id and adresseerbaarobject_id:
                    nummer_and_adresseer_ids["nummeraanduiding_ids"].append(nummeraanduiding_id)
                    nummer_and_adresseer_ids["adresseerbaarobject_ids"].append(adresseerbaarobject_id)
        return nummer_and_adresseer_ids


    def data_for_adressID(self, url: str, _id: str) -> dict:
        params = {"id": _id}
        try:
            response = requests.get(url=url, params=params, timeout=5)
            response.raise_for_status()
            return response.json()
        except requests.RequestException as e:
            print(f"Request failed for {self.address} with error: {e}")
            return {}



    def get_woz_values(self, _ids: List[str]):
        headers = {
            'sec-ch-ua-platform': '"Linux"',
            'Referer': '',
            'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',
            'Accept': 'application/json, text/plain, */*',
            'sec-ch-ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
            'sec-ch-ua-mobile': '?0',
        }
        
        base_url = "https://api.kadaster.nl/lvwoz/wozwaardeloket-api/v1/wozwaarde/nummeraanduiding/"
        woz_values = {year: None for year in range(2014, 2024)}  # 2014 to 2023
        grondoppervlakte = None
        
        with ThreadPoolExecutor() as executor:
            futures = [executor.submit(self.fetch_woz_value, base_url, _id, woz_values) for _id in _ids]
            for future in as_completed(futures):
                result = future.result()
                if result:  # result contains a dictionary with woz_values and grondoppervlakte
                    fetched_woz_values, fetched_grondoppervlakte = result
                    woz_values.update(fetched_woz_values)  # Update woz_values with fetched values
                    if not grondoppervlakte and fetched_grondoppervlakte:
                        grondoppervlakte = fetched_grondoppervlakte  # Set grondoppervlakte if available
        
        return woz_values, grondoppervlakte
    
    def fetch_woz_value(self, base_url, _id, woz_values):
        attempt = 0
        while attempt < 4:
            try:
                response = requests.get(f"{base_url}{_id}", timeout=5)
                if response.status_code == 429:
                    wait_time = 2 ** attempt + random.uniform(0, 1)  # Adds jitter
                    print(f"Rate limited. Waiting for {wait_time:.2f} seconds...")
                    time.sleep(wait_time)
                    attempt += 1
                    continue
                
                response.raise_for_status()  # Raises HTTPError if status is 4xx or 5xx
                data = response.json()
                
                if 'wozWaarden' not in data or not data['wozWaarden']:
                    print(f"No WOZ values found for ID {_id}, skipping.")
                    return None  # Indicate failure to retrieve data
            
                fetched_woz_values = {}
                for item in data['wozWaarden']:
                    year = int(item['peildatum'][:4])
                    if 2014 <= year <= 2023:
                        fetched_woz_values[year] = item['vastgesteldeWaarde']
    
                fetched_grondoppervlakte = data['wozObject'].get('grondoppervlakte', None)
                return fetched_woz_values, fetched_grondoppervlakte  # Return WOZ values and grondoppervlakte
                
            except requests.exceptions.RequestException as e:
                print(f"Request failed for {self.address} with error: {e}")
                return None

    def get_other_features(self, _ids: List[str]):
        headers = {
            'Accept': 'application/json, text/plain, */*',
            'User-Agent': 'Mozilla/5.0',
        }
        params = {'service': 'WFS'}
        features = {}
    
        with ThreadPoolExecutor() as executor:
            futures = [executor.submit(self.fetch_other_feature, _id) for _id in _ids]
            for future in as_completed(futures):
                data = future.result()
                if data:
                    features.update(data)
        return features
     
    def fetch_other_feature(self, _id: str):
            headers = {
                'Accept': 'application/json, text/plain, */*',
                'User-Agent': 'Mozilla/5.0',
            }
            params = {'service': 'WFS'}
            features = {}
    
            data = (
                f'<wfs:GetFeature xmlns:wfs="http://www.opengis.net/wfs" '
                f'service="WFS" version="1.1.0" xsi:schemaLocation="http://www.opengis.net/wfs '
                f'http://schemas.opengis.net/wfs/1.0.0/WFS-transaction.xsd" '
                f'xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" outputFormat="application/json">'
                f'<wfs:Query typeName="verblijfsobject" xmlns:null="http://bag.geonovum.nl">'
                f'<ogc:Filter xmlns:ogc="http://www.opengis.net/ogc">'
                f'<ogc:PropertyIsEqualTo>'
                f'<ogc:PropertyName>identificatie</ogc:PropertyName>'
                f'<ogc:Literal>0{int(_id)}</ogc:Literal>'
                f'</ogc:PropertyIsEqualTo>'
                f'</ogc:Filter>'
                f'</wfs:Query>'
                f'</wfs:GetFeature>'
            )
    
            try:
                response = requests.post(
                    'https://service.pdok.nl/lv/bag/wfs/v2_0',
                    params=params,
                    headers=headers,
                    data=data,
                    timeout=5
                )
                response.raise_for_status()
    
                if response.headers.get("Content-Type") == "application/json":
                    data = response.json()
                    if 'features' in data and data['features']:
                        feature_data = data['features'][0]['properties']
                        return {
                            'gebruiksdoel': feature_data.get('gebruiksdoel'),
                            'bouwjaar': feature_data.get('bouwjaar'),
                            'oppervlakte': feature_data.get('oppervlakte')
                        }
                    
            except requests.exceptions.RequestException as ex:
                print(f"Request failed for ID {self.address} with error: {ex}, peoepepdepde")
            return {}

    
    def to_dataframe(self):
        complete_address = f'{self.postcode} {self.address} {self.woonplaats}'
        ids = self.get_correct_addressID(complete_address)
        
        if not ids:
            return pd.DataFrame()  # Empty DataFrame if no IDs
    
        # Other features
        other_features = self.get_other_features(ids["adresseerbaarobject_ids"])
    
        # Check if "gebruiksdoel" is "woonfunctie"; if not, skip this property
        if other_features.get("gebruiksdoel") != "woonfunctie":
            print(f"Skipping property at {self.address} as 'gebruiksdoel' is not 'woonfunctie'")
            return pd.DataFrame()  # Return an empty DataFrame to indicate skipping
    
        # WOZ values from 2014 to 2023
        woz_values, grondoppervlakte = self.get_woz_values(ids["nummeraanduiding_ids"])
        
        if all(value is None for value in woz_values.values()):
            print(f"No WOZ values available for {self.address}, skipping this property.")
            return pd.DataFrame()
        
        # Prepare data for DataFrame
        data = {
            "address": self.address,
            "bouwjaar": other_features.get("bouwjaar"),
            "grondoppervlakte": grondoppervlakte,
            "oppervlakte": other_features.get("oppervlakte"),
            **{f"woz_{year}": woz_values.get(year) for year in range(2014, 2024)},
        }
    
        self.save_to_csv(pd.DataFrame([data]))
   
    def save_to_csv(self, data: pd.DataFrame):
        file_path = '/home/wouter/Documents/Scriptie/csv/property_data.csv'
        
        if not data.empty:
            data.to_csv(file_path, mode='a', header=not os.path.exists(file_path), index=False)
            print(f"Data for property at {self.address} saved to {file_path}")
        else:
            print(f"No data to save for property at {self.address}")

```
