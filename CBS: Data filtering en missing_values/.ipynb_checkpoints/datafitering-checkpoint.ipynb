## Check number of observations for each 'buurt' 
Aantal observaties moet gelijk zijn aan 11 (van 2014 tot en met 2024) 

Aangezien het onderzoek alleen de buurten betreft, kan de rest uitgesloten worden!




```python
def filter_regions_with_min_observations(df, min_observations=11):

    # Only use the regions annoted with 'BU'
    df_buurt_only = df[df['gwb_code'].str[:2] == 'BU']

    # If "wbi_ind" == 1, group all the buurten based on gwb_code.
    filtered_df = df_buurt_only.groupby(df_buurt_only['gwb_code'].str[2:]).filter(lambda x: len(x) >= min_observations)

    
    # If "wbi_ind" == 2, group based on 'regio' and 'gm_naam' 
    ## Code when this data is avalable

    return filtered_df
    
def standardize_missing_values(df):
    # Define a list of missing value markers, including empty strings
    missing_markers = ['.', '       .', 'NaN', '', None]
    
    df = df.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)

    # Replace these markers with pd.NA
    df = df.replace(missing_markers, pd.NA)


    # Give columns the right type.
    # Exclude columns need to be converted to a string
    exclude_columns = ['regio', 'gm_naam', 'gwb_code']
    for col in df.columns:
        
        # Convert to string
        if col in exclude_columns:
            df[col] = df[col].astype(str)

        # The rest of the columns need to be either a float or integer
        elif df[col].dtype == 'object':
            df[col] = df[col].replace({',': '.', '$': '', '%': ''}, regex=True) #
            df[col] = pd.to_numeric(df[col], errors='coerce')    
    df = df.convert_dtypes()
    return df

filtered_df = filter_regions_with_min_observations(cbs_file_raw)
filt_stand_df = standardize_missing_values(filtered_df)
```

```python
import pandas as pd

def remove_regions_with_missing_woz(df, max_missing=6):

    # Groepeer op 'gwb_code' en tel missende waarden in 'woz'
    missing_counts = df.groupby('gwb_code')['woz'].apply(lambda x: x.isna().sum())

    
    # Selecteer 'gwb_code' met minder dan `max_missing` missende waarden
    valid_gwb_codes = missing_counts[missing_counts < max_missing].index

    filtered_df = df[df['gwb_code'].isin(valid_gwb_codes)]
    
    return filtered_df


def remove_regions_with_low_population(df, min_population=100):

    # df2 = df[['regio','gwb_code', 'a_inw']].copy()
    
    # df2['mean_inw_region'] = df2.groupby('gwb_code')['a_inw'].transform(lambda x: x.fillna(x.mean()))

    # region_population_dict = (
    #     df2.groupby('gwb_code')['mean_inw_region']
    #     .mean()
    #     .reset_index()
    #     .to_dict('records')
    # )

    # for regio in region_population_dict:
    #     if regio['mean_inw_region'] < 60:
    #         print(regio)


    # Identify valid gwb_codes where the mean population is greater than or equal to the threshold
    valid_gwb_codes = df.groupby('gwb_code')['a_inw'].mean()
    valid_gwb_codes = valid_gwb_codes[valid_gwb_codes >= min_population].index
    
    # Filter the DataFrame to include only rows with valid gwb_codes
    filtered_df = df[df['gwb_code'].isin(valid_gwb_codes)]
    
    return filtered_df


filt_woz_df = remove_regions_with_missing_woz(filt_stand_df)
filt_woz_pop_df = remove_regions_with_low_population(filt_woz_df)
display(filt_woz_pop_df)
```

Inzicht krijgen welke colummen we kunnen droppen op basis van aantal missing values (bij zoveel %), en hoe belangrijk de colummen zijn.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_missing_values(df, threshold=20):
    num_rows = len(df)
    
    # Calculate the percentage of missing values per column
    missing_percentages = df.isnull().mean() * 100
    
    # Filter columns where missing percentage is above the threshold
    cols_above_threshold = missing_percentages[missing_percentages > threshold].index.tolist()

    if not cols_above_threshold:
        print(f"No features have more than {threshold}% missing values.")
        return

    cmap = sns.color_palette(["white", "darkslateblue"])  # White for present, DarkSlateBlue for missing


    # Plot heatmap of missing values
    plt.figure(figsize=(12, 6))
    sns.heatmap(df[cols_above_threshold].isnull(), cmap=cmap, cbar=False, yticklabels=False)
    plt.title(f"Missing Values Heatmap (>{threshold}%)", fontsize=14)
    plt.show()

    # Plot bar chart for missing percentages
    plt.figure(figsize=(10, 5))
    missing_percentages[cols_above_threshold].sort_values().plot(kind='barh', color='darkslateblue')
    plt.xlabel("Percentage of Missing Values")
    plt.title(f"Missing Values per Feature (>{threshold}%)")
    plt.show()



# Call the function on your dataset
plot_missing_values(filt_woz_pop_df, threshold=20)
```

```python
import seaborn as sns
import numpy as np

def identify_missing_values(df):

    num_rows = len(df)
    
    # Identify missing features (columns with any missing values)
    missing_features = df.columns[df.isnull().any()].tolist()

    cols_mis_values = []
    
    #print("Detailed Missing Value Report:")
    for column in df.columns:
        missing_count = df[column].isnull().sum()
        if missing_count > 0:
            proc = round((missing_count/num_rows) * 100, 2)
            if proc > 0:
                print(f"Column '{column}' has {missing_count} missing value(s) ----> {proc}%.")
                cols_mis_values.append(column)
    return cols_mis_values
    
cols_mis_values = identify_missing_values(filt_woz_pop_df)

print(cols_mis_values)
```

```python
def count_consecutive_mv(buurt_values: list):
   
    start_index, end_index = None, None
    consecutive_count, highest_count = 0, 0
    temp_start = None  # Temporary start index for the current streak
    
    
    for i in range(len(buurt_values)):
        if pd.isna(buurt_values[i]) and pd.isna(buurt_values[i - 1]):
            consecutive_count += 1
            
            if consecutive_count > highest_count:
                highest_count = consecutive_count
        
        elif pd.isna(buurt_values[i]) and pd.notna(buurt_values[i - 1]):
            consecutive_count = 1

            temp_start = i
            
            if consecutive_count > highest_count:
                highest_count = consecutive_count
                start_index = temp_start

        else:
            
            if pd.isna(buurt_values[i - 1]):
                if consecutive_count == highest_count:
                    end_index = (i - 1)
                    start_index = temp_start

            consecutive_count = 0
                
            
    
    return highest_count, start_index, end_index

# Example usage
data1 = pd.Series([pd.NA, pd.NA, pd.NA, 308, 393, 421, pd.NA, 479, 598, 605])
data1 = pd.Series([pd.NA, pd.NA, 308, pd.NA, pd.NA, pd.NA, 421, pd.NA, 598, 605])

result = count_consecutive_mv(data1.tolist())
print(f"Longest streak: {result[0]}, Start index: {result[1]}, End index: {result[2]}")
```

```python
def linear_interpolation(buurt_data):

    buurt_values = buurt_data.astype('Float64')

    
    while buurt_values.isnull().sum() > 0:
        
        # Extrapolation: wanneer er een missing value aan het begin is.
        if pd.isna(buurt_values.iloc[0]) and pd.notna(buurt_values.iloc[-1]):

            # Interpolation eerst: Als er een missing value is voor buurt_values[2]
            if pd.isna(buurt_values.iloc[2]):
                buurt_values = buurt_values.interpolate(method='linear', limit_direction='forward')
          
            slope = buurt_values.iloc[2] - buurt_values.iloc[1]
            buurt_values.iloc[0] = buurt_values.iloc[1] - slope

        
        # Extrapolation: wanneer er een missing value aan het einde is.
        elif pd.isna(buurt_values.iloc[-1]) and pd.notna(buurt_values.iloc[0]):
            
            # Interpolation eerst: Als er een missing value is voor buurt_values[-3]
            if pd.isna(buurt_values.iloc[-3]):
                buurt_values = buurt_values.interpolate(method='linear', limit_direction='backward')
            
            slope = (buurt_values.iloc[-2] - buurt_values.iloc[-3])
            buurt_values.iloc[-1] = buurt_values.iloc[-2] + slope


        # Extrapolation: wanneer er een missing value aan het begin en einde is.
        elif pd.isna(buurt_values.iloc[0]) and pd.isna(buurt_values.iloc[-1]):

            # Interpolation eerst: Als er een missing value is voor buurt_values[2] en buurt_values[-3]
            if pd.isna(buurt_values.iloc[2]) or pd.isna(buurt_values.iloc[-3]):
                interpolated_subset = buurt_values.iloc[1:-1].interpolate(method='linear', limit_direction='forward')
                buurt_values.iloc[1:-1] = interpolated_subset

            slope1 = (buurt_values.iloc[2] - buurt_values.iloc[1])
            buurt_values.iloc[0] = buurt_values.iloc[1] - slope1
            slope2 = (buurt_values.iloc[-2] - buurt_values.iloc[-3])
            buurt_values.iloc[-1] = buurt_values.iloc[-2] + slope2
            
        
        # Anders alleen interpolation.
        else:
            buurt_values = buurt_values.interpolate(method='linear', limit_direction='both')

    return buurt_values

data1 = pd.Series([pd.NA, 2200, 2270, 2270, 2310, 2510, 2860, 3010, 3050, pd.NA])

linear_interpolation(data1)
```

```python
import pandas as pd

def weighted_slope(start_value, end_value, start_index, end_index):
    # # Adjust weights to favor nearer points
    # weight_start = 1 / (start_index + 1)  # Higher weight for closer points
    # weight_end = 1 / (end_index + 1)     # Higher weight for closer points

    # # Calculate weighted slope
    # slope = (end_value - start_value) / (end_index - start_index)
    # weighted_slope = slope * ((weight_start + weight_end) / 2)

    weighted_slope = (end_value - start_value) / (end_index - start_index)

    return weighted_slope

def consecutive_extrapolation(buurt_data, start_index, end_index):

    buurt_values = pd.Series(list(buurt_data))
    # Make sure right data type, improve 
    buurt_values = buurt_values.astype('Float64')

    smal_known_i = buurt_values.first_valid_index()
    larg_known_i = buurt_values.last_valid_index()
    
    ## Compute slope between first and last known point
    #w_slope = weighted_slope(buurt_values.iloc[smal_known_i], buurt_values.iloc[larg_known_i], smal_known_i, larg_known_i)
    
    # When the consecutive missing values occur at the beginning, begin iterating at the end.
    if start_index == 0:
        loop_start = len(buurt_values) - 1
        loop_end = start_index - 1
        direction = -1
    
    # When the consecutive missing values occur at the end or in between, begin iterating at the start.
    elif start_index != 0:
        loop_start = 0
        loop_end = len(buurt_values)
        direction = 1
        

    for i in range(loop_start, loop_end, direction):

        valid_indices = buurt_values[buurt_values.notna()].index

        if pd.isna(buurt_values.iloc[i]):  # Only extrapolate missing values
            
            prev_index = valid_indices[valid_indices < i]
            next_index = valid_indices[valid_indices > i]

            if not prev_index.empty and not next_index.empty: 

                
                prev_known_index = prev_index.max()
                prev_known_value = buurt_values.iloc[prev_known_index] 
                next_known_index = next_index.min()
                next_known_value = buurt_values.iloc[next_known_index]

                slope = weighted_slope(prev_known_value, next_known_value, prev_known_index, next_known_index)
                #print(slope)

                if start_index == 0:
                    buurt_values.iloc[i] = buurt_values.iloc[i + 1] - slope
                    #buurt_values.iloc[i] = next_known_value + slope * (i - next_known_index)

                else: 
                    buurt_values.iloc[i] = buurt_values.iloc[i - 1] + slope
                    #buurt_values.iloc[i] = prev_known_value + slope * (i - prev_known_index)

            
            elif prev_index.empty and not next_index.empty: 
                next_known_index = next_index.min()
                next_known_value = buurt_values.iloc[next_known_index]
                
                second_known_index = next_index[1]                
                second_known_value = buurt_values.iloc[second_known_index]

                slope = weighted_slope(next_known_value, second_known_value, next_known_index, second_known_index)
                
                buurt_values.iloc[i] = buurt_values.iloc[i+1] - slope


            elif not prev_index.empty and next_index.empty: 
                prev_known_value = buurt_values.iloc[prev_index.max()]
                
                second_known_index = prev_index[len(prev_index)-2]
                second_known_value = buurt_values.iloc[second_known_index]

                slope = weighted_slope(second_known_value, prev_known_value, second_known_index, prev_index.max())

                buurt_values.iloc[i] = buurt_values.iloc[i-1] + slope 
                

    return buurt_values


# # verspreide huizen veendam [203, 205, 205, 211, 254, 276, 278, 292, 327, 353]
data1 = pd.Series([203, pd.NA, 205, 211, pd.NA, pd.NA, pd.NA, 292, 327, 353])

#data3 = pd.Series([87, 87, 88, 94, 103, pd.NA, pd.NA, 163, 177, 181])

consecutive_extrapolation(data1, 4, 6)
#print(list(result))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

def spline_interpolation(buurt_data):

    buurt_values = buurt_data.astype('Float64')
    
    x = np.arange(len(buurt_values))
    
    # Create a mask for the valid (non-NaN) values
    mask = ~buurt_values.isna()
    
    # Fit cubic spline to the valid data points
    cs = CubicSpline(x[mask], buurt_values[mask], bc_type='natural', )
    
    # Generate new x values for interpolation
    x_new = np.linspace(x.min(), x.max(), 100)
    y_new = cs(x_new)
    
    # Interpolate the missing values in the original data
    buurt_values[buurt_values.isna()] = cs(x[buurt_values.isna()])
    
    # Adjust the first few values to ensure they are lower than subsequent values

    return buurt_values

# real values [299, 310, 315, 313, 217, 226, 233, 258, 316, 337]
data1 = pd.Series([np.nan, np.nan, 315, np.nan, 217, np.nan, np.nan, 258, 316, 337])  
imputed_values = spline_interpolation(data1)

print(imputed_values)
```

```python
### import matplotlib.pyplot as plt
from itertools import islice
from scipy.stats import linregress


def linearity(df):
    # Create a new DataFrame to store the results and copy the first 5 rows entirely from the original DataFrame
    new_df = complete_filtered_df[['regio', 'gm_naam', 'gwb_code', 'ind_wbi', 'year']].copy()   

    # Select the colomns
    columns = df.columns

    for col in columns[5:]:
        new_df[col] = None   # Voeg nieuwe colomn toe aan df
        print(f'Processing column: {col}')

        # Bereken percentage missing values per colomn, kan iets mee worden gedaan
        missing_count = df[col].isnull().sum()        
        if missing_count > 0:
            proc = round((missing_count/len(df)) * 100, 2)

        # Als column geen missing values heeft kan colom worden toegevoegd aan new_df
        else:
            new_df[col] = df[col]
            continue

        # Groepeer buurten om er door te itereren en per colom de ontbrekende waarde te vinden
        buurten = df.groupby('gwb_code')[col]
        
        #for buurt in islice(buurten, 300):
        for buurt_code, buurt_data in buurten:

            row_index = buurt_data.index

            if buurt_data.isnull().any():
                x = buurt_data.dropna().index
                y = buurt_data.dropna()

                

            # Get row index buurt_data
            row_index = buurt_data.index
            #print(row_index)

            if buurt_data.isnull().any():                
                # identife x and y, for linregression
                x = buurt_data.dropna().index
                y = buurt_data.dropna()

                # Als er voor de buurt geen data is ga door, bedenk wat anders voor later
                if (x.empty or y.empty) or (len(buurt_data[buurt_data.isnull()]) >= 7): 
                    new_df.loc[row_index, col] = buurt_data
                    #print(buurt_data)
                    #print(buurt_code, y)
                    
                else:   
                    slope, intercept, r_value, p_value, std_err = linregress(x, y)
                    r_sqrt = r_value**2      

                    # Wanneer er 1 missing value is in de buurt
                    if len(buurt_data[buurt_data.isnull()]) == 1: 
                        if r_sqrt > 0.75:
                            handled_data = linear_interpolation(buurt_data) 

                        # Als r-waarde kleiner of gelijk is aan 75%: Gebruik .... voor ontbrekende waarden
                        elif r_sqrt <= 0.75:
                            handled_data = spline_interpolation(buurt_data)                          
                            #for original, handled in zip(buurt_data, handled_data):
                                #print(f"Case2: {str(original):<15} {str(handled):<15}")
       
                    
                    # There are multiple missing_values
                    else:
                        consecutive_coount, start_index, end_index = count_consecutive_mv(list(buurt_data))
    
                        # When there arent any consecutive missing values
                        if consecutive_coount == 1:
                            
                            if r_sqrt > 0.75:
                                handled_data = linear_interpolation(buurt_data)

                            elif r_sqrt <= 0.75:
                                handled_data = spline_interpolation(buurt_data)
                                
                        elif consecutive_coount > 1:
                            if r_sqrt > 0.75:
                                handled_data = consecutive_extrapolation(buurt_data, start_index, end_index)
                                
                            elif r_sqrt <= 0.75:
                                handled_data = spline_interpolation(buurt_data)  
                                
                
                # Make sure handled data has same dtype and index as buurt_data
                handled_data = handled_data.astype(buurt_data.dtype)
                handled_data.index = buurt_data.index 
                
                # Update the new DataFrame with handled data
                new_df.loc[row_index, col] = handled_data

            else:
                # If no missing values, copy the original data
                new_df.loc[row_index, col] = buurt_data


    return new_df
                    

new_df = linearity(complete_filtered_df)


```
